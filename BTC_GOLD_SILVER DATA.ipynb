```python
import pandas as pd
from openbb_terminal.sdk import openbb
```

```python
gold_monthly = openbb.stocks.load(symbol = 'gc=f', start_date = '2014-09-29', end_date = '2023-09-18', monthly = True)
```

```python
gold_monthly
```

```python
gold_monthly2 = gold_monthly[["Close", "Volume"]]
```

```python
gold_monthly2.reset_index(inplace = True)
```

```python
gold_new = gold_monthly2.rename(columns={'date': 'Month_Year', 'Close': 'Closing_Price'})
```

```python
gold_new
```

```python
gold_new.to_csv("currency_csvs/gold.csv")
```
